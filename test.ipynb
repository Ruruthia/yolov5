```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
loss_wh = pd.read_csv('loss_wh.txt', header=None)
loss_xy = pd.read_csv('loss_xy.txt', header=None)
loss_cls = pd.read_csv('loss_cls.txt', header=None)
loss_obj = pd.read_csv('loss_obj.txt', header=None)
acc = pd.read_csv('acc.txt', sep=" ", header=None).to_numpy()
```

```python
plt.plot(loss_xy)
```

```python
plt.plot(loss_wh)
```

```python
plt.plot(loss_cls)
```

```python
plt.plot(loss_obj)
```

```python
loss = loss_cls + loss_wh + loss_xy + loss_obj
plt.plot(loss)
```

```python
loss_per_epoch = loss.to_numpy().reshape(-1, 8).mean(axis=1)
plt.plot(loss_per_epoch)
```

```python
thresholds = [0.5, 0.6, 0.7, 0.8, 0.9]
for i, threshold in enumerate(thresholds):
    plt.plot(acc[:, i], label=threshold)
plt.legend()
```

```python
acc_per_epoch = acc.reshape(-1, 8, 5).mean(axis=1)
for i, threshold in enumerate(thresholds):
    plt.plot(acc_per_epoch[:, i], label=threshold)
plt.legend()
```
